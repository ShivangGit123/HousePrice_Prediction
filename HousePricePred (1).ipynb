```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Dense , Dropout
from keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
df
```

```python
df['ocean_proximity'].value_counts()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['ocean_proximity'].value_counts()
```

```python
df['ocean_proximity'] = df['ocean_proximity'].map({'<1H OCEAN':0 , 'INLAND' :1 , 'NEAR OCEAN' :2 , 'NEAR BAY' :3 , 'ISLAND' :4})
```

```python
df['ocean_proximity'].value_counts()
```

```python
df['ocean_proximity'].head()
```

```python
df
```

```python
X = df.drop('median_house_value' ,axis=1)
y = df['median_house_value']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=1)
```

# ***Scaling***

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

# ***Model Training***

```python
model = Sequential()
model.add(Dense(1000 , input_shape=[X_train.shape[1]] , activation='relu'))
model.add(Dense(500 , activation='relu'))
model.add(Dense(250 , activation='relu'))
model.add(Dense(1 , activation='linear'))
model.compile(optimizer='adam' , loss='mse')
```

```python
callback = EarlyStopping(
    monitor = 'val_loss',
    min_delta = 0.00001 ,
    patience = 20 ,
    verbose = 1,
    mode = "min" ,
    baseline = None ,
    restore_best_weights = True
)

model.fit(X_train , y_train , validation_split=0.2 , epochs=30 , callbacks=[callback] , batch_size=32)
```

# ***Results***

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_squared_log_error

y_pred = model.predict(X_test)
print("mae :", mean_absolute_error(y_test,y_pred))
print("mse :", mean_squared_error(y_test,y_pred))
print("mae :", mean_squared_log_error(y_test,y_pred))
print("score :", r2_score(y_test,y_pred))
```

# ***Prediction function***

```python
def pred(longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity):
    features = np.array([longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity])
    features_scaled = scaler.transform([features])
    results = model.predict(features_scaled)
    return results[0]
```

```python
import numpy as np
import gradio as gr
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.models import load_model
from tensorflow.keras.losses import MeanSquaredError

model = load_model("model_ann.h5", custom_objects={'mse': MeanSquaredError()})


# Load model and scaler
model = load_model("model_ann.h5" , custom_objects={'mse': MeanSquaredError()})
scaler = pickle.load(open("scaler.pkl", "rb"))

# Map labels to numeric values
ocean_map = {
    '<1H OCEAN': 0,
    'INLAND': 1,
    'NEAR OCEAN': 2,
    'NEAR BAY': 3,
    'ISLAND': 4
}

# Reverse map for UI dropdown (optional, if needed)
ocean_labels = list(ocean_map.keys())

# Define prediction function
def predict_price(longitude, latitude, housing_median_age, total_rooms,
                  total_bedrooms, population, households, median_income, ocean_proximity):
    try:
        ocean_encoded = ocean_map[ocean_proximity]  # Convert label to number

        # Create input array
        input_data = np.array([[longitude, latitude, housing_median_age,
                                total_rooms, total_bedrooms, population,
                                households, median_income, ocean_encoded]])

        # Ensure values are within training range
        input_data = np.clip(input_data, scaler.data_min_, scaler.data_max_)

        # Scale input
        input_scaled = scaler.transform(input_data)

        # Predict
        prediction = model.predict(input_scaled)[0][0]
        return f"ðŸ  Predicted House Value: ${prediction:,.2f}"

    except Exception as e:
        return f"âŒ Error: {str(e)}"

# Gradio interface
inputs = [
    gr.Number(label="Longitude"),
    gr.Number(label="Latitude"),
    gr.Number(label="Housing Median Age"),
    gr.Number(label="Total Rooms"),
    gr.Number(label="Total Bedrooms"),
    gr.Number(label="Population"),
    gr.Number(label="Households"),
    gr.Number(label="Median Income"),
    gr.Dropdown(choices=ocean_labels, label="Ocean Proximity")
]

output = gr.Textbox(label="Prediction")

# Launch interface
gr.Interface(fn=predict_price, inputs=inputs, outputs=output, title="California House Price Predictor").launch()
```
